```python
class Vertex:
    def __init__(self, node):
        self.id = node
        self.adjacent = {}
        self.obstaclecnt = 0
        self.recipe = False
        self.morid = False
        self.parent = None

    def __str__(self):
        return str(self.id) + ' adjacent: ' + str([x.id for x in self.adjacent])

    def add_neighbor(self, neighbor, weight=0):
        self.adjacent[neighbor] = weight

    def get_connections(self):
        return self.adjacent.keys()  

    def get_id(self):
        return self.id

    def get_weight(self, neighbor):
        return self.adjacent[neighbor]

    def get_parent(self):
        return self.parent

    



class Graph:
    def __init__(self):
        self.vert_dict = {}
        self.num_vertices = 0

    def __iter__(self):
        return iter(self.vert_dict.values())

    def add_vertex(self, node):
        self.num_vertices = self.num_vertices + 1
        new_vertex = Vertex(node)
        self.vert_dict[node] = new_vertex
        return new_vertex

    def get_vertex(self, n):
        if n in self.vert_dict:
            return self.vert_dict[n]
        else:
            return None

    def add_edge(self, frm, to, cost = 0):
        if frm not in self.vert_dict:
            self.add_vertex(frm)
        if to not in self.vert_dict:
            self.add_vertex(to)

        self.vert_dict[frm].add_neighbor(self.vert_dict[to], cost)
        self.vert_dict[to].add_neighbor(self.vert_dict[frm], cost)

    def get_vertices(self):
        return self.vert_dict.keys()
    
    def sethard(self,n):
        self.vert_dict[n].obstaclecnt =  self.vert_dict[n].obstaclecnt + 1



class Queue:
	def __init__(self):
		self.datalist = []

	def enqueue(self, data):
		self.datalist.append(data)

	def dequeue(self):
		if self.is_empty() == True:
			return
		data = self.datalist[0]
		self.datalist.pop(0)
		return data

	def is_empty(self):
		if len(self.datalist) == 0:
			return True
		return False
```

```python
def BFS(head,targets):
    pq = Queue()
    explored = set()
    pq.enqueue(head)
    while not pq.is_empty():
        v = pq.dequeue()
        explored.add(v)
        for item in v.get_connections():
            if item not in explored:
                if item.parent == None:
                    item.parent = v
                pq.enqueue(item)
                if item.get_id() in targets:
                    return item
        

    
           
#def printPath(root):
#    if root.parent == None:
#        print(root.get_id())
#        return#
#   printPath(root.get_parent())
#    print(root.get_id())

def getPath(root):
    path = []
    while  True:
        path.append(root.get_id())
        root = root.get_parent()
        if root == None:
            break
    path.reverse()
    return path
        


def clearPaths(Graph):
    for v in Graph:
        v.parent = None

def moridCheck(targets, collected, goals):
    for morid in targets:
        if(set(targets[morid]).issubset(set(collected))):
            goals.append(morid)

            

```
